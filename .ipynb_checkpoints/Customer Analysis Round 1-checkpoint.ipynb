```python
data.dtypes
```

 - **Numerical columns, continuous**:income,monthly_premium_auto, total_claim_amount, customer_lifetime_value are continuous, because they  can take on any value within a specified range.
 - **Numerical columns,Discrete** - number_of_open_complaints,because it can only take on specific values within a defined range or set. These values are  whole numbers and cannot be further subdivided.
 - **Categorical, nominal**: state and gender,because they have no inherent order or ranking.
 -  **Categorical, ordinal**: education, policy_type, vehicle_class - have  inherent order or ranking.
 - **Customer** is not a numerical or categorical, it's a unique identifier of customer.

 - **customer** - is a inique identifier for each customer                   
 - **st** - geographic location of customer                      
 - **gender** - customer's gender                      
 - **education** - customer's level of education                 
 - **customer_lifetime_value** - typically represents the total amount of money a customer is expected to spend throughout their relationship with the company. 
 - **income** - customer's income level or yearly income.
 - **monthly_premium_auto** - the amount the customer pays each month for their insurance premium related to their vehicle.        
 - **number_of_open_complaints** - number of complaints currently open by the customer.    
 - **policy_type** - type of insurance policy held by the customer (e.g., Comprehensive, Collision, Liability, etc.).             
 - **vehicle_class** - Class or type of vehicle insured (e.g., SUV, Sedan, Sports Car, etc.).  
 - **total_claim_amount** - the total amount claimed by the customer for insurable events.
